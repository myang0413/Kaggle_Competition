## Importing Necessary Datasets

```python
#Import training dataset
train = pd.read_csv("/kaggle/input/walmart-recruiting-store-sales-forecasting/train.csv.zip")
train
```

```python
#Importing test dataset
test = pd.read_csv("/kaggle/input/walmart-recruiting-store-sales-forecasting/test.csv.zip")
test
```

## Data Pre-processing

```python
#Change Date Column to Date Format and splitting it to Year/Month/Day Columns on Train Dataset
train["Date"]=pd.to_datetime(train["Date"])
train["Year"]=train["Date"].dt.year
train["Month"]=train["Date"].dt.month
train["Day"]=train["Date"].dt.day
```

```python
#Delete categorical variables
train_2=train.drop(columns=["Date","Weekly_Sales"])
train_2
```

```python
#Change Date Column to Date Format and splitting it to Year/Month/Day Columns on Test Dataset
test["Date"]=pd.to_datetime(test["Date"])
test["Year"]=test["Date"].dt.year
test["Month"]=test["Date"].dt.month
test["Day"]=test["Date"].dt.day
```

```python
test_2=test.drop(columns=["Date"])
test_2
```

## Modelling (Random Forest) 

```python
#Importing Necessary Package
from sklearn.ensemble import RandomForestRegressor

#Creating and Fitting the Random Forest model based on train and test dataset
#Assigning n_jobs for the faster execution of the model 
rf=RandomForestRegressor(n_jobs=4)
rf.fit(train_2,train["Weekly_Sales"])
result=rf.predict(test_2)
result
```

## Submission 

```python
sub=pd.read_csv("/kaggle/input/walmart-recruiting-store-sales-forecasting/sampleSubmission.csv.zip")
sub
```

```python
sub["Weekly_Sales"]=result
sub
```
