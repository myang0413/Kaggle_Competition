## Importing datasets and Data Pre-processing

```python
#Import train dataset
train = pd.read_csv("/kaggle/input/bike-sharing-demand/train.csv")
train
```

```python
#Check the distribution of the outcome variable
import seaborn as sns
sns.displot(train["count"])
```

```python
#See if log transformation of the outcome variable can change the distribution close to normal distribution
sns.displot(np.log(train["count"]))
```

```python
#Import test dataset
test = pd.read_csv("/kaggle/input/bike-sharing-demand/test.csv")
test

```

```python
#Only add Hours, Year, and Weekday columns as those are the important columns to consider to train dataset
#Month and Day columns would not be necessary to add to the train dataset as train dataset counts days from 1st to 19th day of the month and test dataset only counts from 20th to last day of the month
#Month would be closely linked to the day so it would not be necessary to add it 
train["datetime"]=pd.to_datetime(train["datetime"])
train["Year"]=train["datetime"].dt.year
train['Hours']= train['datetime'].dt.hour
train['Weekday']= train['datetime'].dt.weekday
```

```python
train_2=train.drop(columns=["datetime","count","casual","registered"])
train_2
```

```python
test["datetime"]=pd.to_datetime(test["datetime"])
test["Year"]=test["datetime"].dt.year
test['Hours']= test['datetime'].dt.hour
test['Weekday']= test['datetime'].dt.weekday
```

```python
test_2=test.drop(columns=["datetime"])
test_2
```

## Random Forest Modelling

```python
#Create a random forest model
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(n_jobs=4)

#Fitting the model and transforming the logarithm 
rf.fit(train_2,np.log(train["count"]))
result=rf.predict(test_2)
result
```

## Submission

```python
sub_3=pd.read_csv("/kaggle/input/bike-sharing-demand/sampleSubmission.csv")
sub_3
```

```python
#Convert the count value of the result from the prediction to original level by taking with exponential function
sub_3["count"]=np.exp(result)
sub_3
```
